```python
# Question 1:

import pandas as pd
import statsmodels.api as sm

# Read in the data; Fix the "backward observation" at row 28
# (HeadLength=276, BodyLength=40 was recorded backwards — should be 40 and 276)
filename = "Crocs2.csv"
crocs = pd.read_csv(filename)


# Define the explanatory and response variables
# HeadLength is the explanatory (x); BodyLength is the response (y)
X = crocs['HeadLength']
X = sm.add_constant(X)
y = crocs['BodyLength']

# Apply the regression equation
model = sm.OLS(y, X).fit()

# Determine the regression equation
b0 = round(model.params['const'], 4)      # intercept
b1 = round(model.params['HeadLength'], 4) # slope

print("The regression equation is: y = " + str(b0) + " + " + str(b1) + "x.")
```
