```python
#Python Code (3 / 4)

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

df = pd.read_csv("fastfood.csv")

# Scatterplot
sns.regplot(x="Restaurants", y="Obesity", data=ff, ci=0)
plt.title("Fast Food Restaurants per Capita vs. Obesity Rate")
plt.xlabel("Fast Food Restaurants per Capita")
plt.ylabel("Obesity Rate (%)")

# Correlation
r = df[['Restaurants','Obesity']].corr().iloc[0,1]
r = round(r, 4)
print("The correlation is " + str(r) + ".")
```

```python
#Python Code (6, 7, 10)

X = sm.add_constant(ff['Restaurants'])
y = ff['Obesity']
model = sm.OLS(y, X).fit()

b0 = round(model.params['const'], 4)
b1 = round(model.params['Restaurants'], 4)
r2 = round(model.rsquared, 4)

print("Regression equation: y = " + str(b0) + " + " + str(b1) + "x")
print("R-squared = " + str(r2))
```

```python
# Question 4: Basketball

P_H = 0.80
P_M = 0.20

# Part 1: P(M)
print("Part 1: P(M) =", P_M)

# Part 2: Two free throws
print("\nPart 2: Two free throws")
print("Sample space: {HH, HM, MH, MM}")
P_HH = P_H * P_H
P_MM = P_M * P_M
P_one_hit = P_H*P_M + P_M*P_H
print("P(HH) =", P_HH)
print("P(MM) =", P_MM)
print("P(exactly one hit) =", P_one_hit)

# Part 3: Three free throws
print("\nPart 3: Three free throws")
print("Sample space: {HHH, HHM, HMH, MHH, HMM, MHM, MMH, MMM}")
P_HHH = P_H**3
print("P(HHH) =", P_HHH)
# First miss on third attempt: HHM
P_HHM = P_H * P_H * P_M
print("P(first miss on 3rd attempt, i.e., HHM) =", P_HHM)
# At least one miss = complement of zero misses
P_at_least_one_miss = 1 - P_HHH
print("P(at least one miss) = 1 - P(HHH) =", P_at_least_one_miss)
```
