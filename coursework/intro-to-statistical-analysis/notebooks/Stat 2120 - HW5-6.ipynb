```python
#Question 1, Part 1
#Part 1 — P(single game score ≥ 70)
import scipy.stats as stats
import numpy as np

mu, sigma = 68.6, 4
P = 1 - stats.norm.cdf(70, mu, sigma)
print(P)  # 0.3632
```

```python
#Part 2 — P(average over 6 games > 70)
sigma_xbar = 4 / np.sqrt(6)
P = 1 - stats.norm.cdf(70, mu, sigma_xbar)
print(P)  # 0.1956
```

```python
#Part 3 — P(total points over 6 games < 415)
P = stats.norm.cdf(415/6, mu, sigma_xbar)
print(P)  # 0.6357
```

```python
#Problem 2 — Cholesterol Confidence Interval (n = 37, σ = 47.7, x̄ = 251.89)
#Part 1 — 99% Confidence Interval

import pandas as pd

chol = pd.read_csv('cholesterol.csv')
xbar = chol['level'].mean()   # 251.8919
n = len(chol)                  # 37
sigma = 47.7

z_star = stats.norm.ppf(0.995)  # 2.5758
ME = z_star * sigma / np.sqrt(n)
LL = xbar - ME
UL = xbar + ME
print(f"({LL:.4f}, {UL:.4f})")  # (231.6927, 272.0911)
```

```python
#2.5 (Part 5 — Margin of Error)
ME = z_star * sigma / np.sqrt(n)
print(ME)  # 20.1992
```

```python
#Part 6 — Sample size for total width = 20 (margin of error = 10)
m = 10
n_needed = np.ceil((z_star * sigma / m)**2)
print(n_needed)  # 151.0
```

```python
#Part 3: Test stat / P-value for Hypothesis (pop mean > 225)
mu0 = 225
z = (xbar - mu0) / (sigma / np.sqrt(n))
p_val = 1 - stats.norm.cdf(z)
print(f"z = {z:.4f}")    # 3.4293
print(f"p = {p_val:.4f}") # 0.0003
```
